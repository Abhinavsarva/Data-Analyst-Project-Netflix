```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

 Missing Value & Outlier check (Treatment optional)
 Outliers: We've used boxplots to detect outliers in numerical columns like 'Release_year' and 'Duration'. We can handle these outliers by clipping them to a defined range or removing the rows containing outliers.


```python
sns.boxplot(x=net['release_year'])
sns.boxplot(y=net['duration'])
plt.show()

```

 Value Counts and Unique Attributes
 Non-Graphical Analysis: Value counts and unique attributes

```python
unique_directors = net['director'].nunique()
unique_countries = net['country'].nunique()
unique_genres = net['listed_in'].nunique()
unique_ratings = net['rating'].nunique()
```

```python
(unique_countries), (unique_genres), (unique_ratings), (unique_directors)
```

Visual Analysis - Univariate, Bivariate after pre-processing of the data

```python
#Distplot
sns.distplot(net['release_year'].dropna(), kde=True)
plt.title('distribution of release_year')
plt.show()
```

```python
#countplt
sns.countplot(net['release_year'])
plt.title('countplot of release_year')
plt.show()
```

```python
net['duration'] = net['duration'].str.extract('(\d+)').astype(float)
plt.hist(net['duration'].dropna(), bins=20)
plt.title('Histogram of duration')
plt.xlabel('duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
